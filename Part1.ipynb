# A)

### 1. Let X be a random variable with probability density function.
### f(x) = (θ + 1)(1 − x)θ, 0 ≤ x ≤ 1, θ > 0

#### Enter an desired value for θ 

```python
dtheta = int(input("Enter an desired value for θ : "))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def f(x, theta):
    return (theta + 1) * (1 - x)**theta

def cdf(x, theta):
    return quad(f, 0, x, args=(theta))[0]

theta_val = dtheta
x_vals = np.linspace(0, 1, 1000)
y_vals = [cdf(val, theta_val) for val in x_vals]

print("Fx[0.25] = ", cdf(0.25, theta_val))
plt.plot(x_vals, y_vals)
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Cumulative Density Function (CDF)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def f(x, theta):
    return (theta + 1) * (1 - x)**theta

def cdf(x, theta):
    return quad(f, 0, x, args=(theta))[0]

theta_val = 2  
x_vals = np.linspace(0, 1, 1000)
y_vals = [cdf(val, theta_val) for val in x_vals]

specific_x = 0.25
specific_y = cdf(specific_x, theta_val)  
print("Fx[0.25] = ", specific_y)

plt.plot(x_vals, y_vals)
plt.plot(specific_x, specific_y, 'ro')  
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Cumulative Density Function (CDF)')
plt.show()
```
