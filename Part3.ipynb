### Using python, Construct a plot for the probability mass function and the cumulative probability distribution of a binomial random variable Bin(n = 8, p = 0.3). Also using python, Find the smallest value of k such that P(X ≤ k) ≥ 0.44 when X ∼ Bin(n = 8, p = 0.7). Calculate P(Y ≥ 3) if Y ∼ Bin(20, 0.2).

# A)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 8
p = 0.3
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
cdf = binom.cdf(x, n, p)


plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(x, pmf)
plt.title('Probability Mass Function (n=8, p=0.3)')
plt.xlabel('x')
plt.ylabel('P(X=x)')

plt.subplot(1, 2, 2)
plt.step(x, cdf)
plt.title('Cumulative Distribution Function (n=8, p=0.3)')
plt.xlabel('x')
plt.ylabel('P(X<=x)')

plt.show()
```

# B)

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 8
p = 0.7
k = 0
cdf_value = 0

while cdf_value < 0.44:
    k += 1
    cdf_value = binom.cdf(k, n, p)

print("The smallest value of k such that P(X ≤ k) ge 0.44 is:", k)
```
